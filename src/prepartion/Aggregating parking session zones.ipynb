```python
import pandas as pd
```

```python
# importing data sets

parkingsessions = pd.read_csv("data/parking_sessions_ev_hv.csv")
```

```python
# check if it is dataframe
isinstance(parkingsessions, pd.DataFrame)

parkingsessions
```

```python
# take subset of what's important
parking = parkingsessions[["sessionid", "zoneid","latitude","longitude", "sessionduration"]]

parking
```

```python
# aggregating by zones using "pandas.DataFrame.groupby"

# df_municipalities = df.groupby(["Gemeentecode", "Datum"], as_index=False).agg({
#     "PositiefGetest": "mean",
#     "Gemeente": "first",
# })

parking_aggregated = parking.groupby(["zoneid"], as_index=False).agg({
    "sessionid": "count",
    "latitude": "first",
    "longitude": "first",
    "sessionduration": "mean",
})

parking_aggregated
```

```python
# changing names
parking_aggregated.rename(columns={'sessionid': 'session_count', 'sessionduration': 'session_average'}, inplace=True)

parking_aggregated
```

```python
# exporting back to csv
parking_aggregated.to_csv("data/parking_aggregated.csv")
```
