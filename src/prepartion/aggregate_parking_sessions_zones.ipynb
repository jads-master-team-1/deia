```python
import pandas as pd
```

```python
# Importing data sets
parking_sessions = pd.read_csv("../../data/parking_sessions_ev_hv.csv")
```

```python
# Check if it is dataframe
isinstance(parking_sessions, pd.DataFrame)

parking_sessions
```

```python
parking_sessions["startofsession"].min(), parking_sessions["startofsession"].max()
```

```python
parking_sessions["endofsession"].min(), parking_sessions["endofsession"].max()
```

```python
# Take subset of what's important
parking_sessions = parking_sessions[["sessionid", "zoneid","latitude","longitude", "sessionduration"]]

parking_sessions
```

```python
# Aggregating by zones using "pandas.DataFrame.groupby"
parking_aggregated = parking_sessions.groupby(["zoneid"], as_index=False).agg({
    "sessionid": "count",
    "latitude": "first",
    "longitude": "first",
    "sessionduration": "mean",
})

parking_aggregated
```

```python
# Changing name
parking_aggregated.rename(columns={
    "sessionid": "session_count",
    "sessionduration": "session_average",
}, inplace=True)

parking_aggregated
```

```python
# Exporting back to csv
parking_aggregated.to_csv("../../data/parking_aggregated.csv", index=False)
```
